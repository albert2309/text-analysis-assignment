```python
from nltk.tokenize import sent_tokenize, word_tokenize

# Sentence Segmentation
sampleText = "This programme is designed to provide students with knowledge and applied skills in data science, big data analytics, and business intelligence. It aims to develop analytical and investigative knowledge and skills using data science tools and techniques, and to enhance data science knowledge and critical interpretation skills. Students will understand the impact of data science up modern processes and business, be able to identify, and implement specific tools, practices, features, and techniques to enhance the analysis of data."

Sentences = sent_tokenize(sampleText)

print("There are ", len(Sentences), "sentencs in text\n")

counter = 0
for sent in Sentences:
    counter+=1
    print(counter, ".", sent, "\n")
```

```python
#Removing Stop Words and punctuation
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string 
fileHandler = open('sample01.txt')
sampleText = fileHandler.read()

Tokens = word_tokenize(sampleText)
print("There are ", len(Tokens), "words in this text\n")

stopTokens = stopwords.words("english") + list(string.punctuation)
filteredTokens =[]
for w in Tokens:
    if w not in stopTokens: 
        filteredTokens.append(w)
        
print("There are ", len(filteredTokens), "words in this text after removing stop words and punctuation\n")
print(filteredTokens)
```
