```python
#implement porter stemmer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

ps = PorterStemmer()

example_words = ["python","pythoner","pythoning","pythoned","pythonly"]

for w in example_words:
    print(ps.stem(w))
```

```python
#Tokenize word
new_text = """It is very important to be pythonly while you are pythoning with python. All pythoners have pythoned poorly at least once."""

words = word_tokenize(new_text)

for w in words:
    print(ps.stem(w))
```

```python
#Tokenize and stemming using three different stemmers (Porter, Lancaster, and WordNet)
import nltk
raw = """DENNIS: Listen, strange women lying in ponds distributing swords
is no basis for a system of government. 
Surpeme executive power derives from a mandate from the masses, 
not from some farcical aquatic ceremony."""

tokens = word_tokenize(raw)

porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
wnl = nltk.WordNetLemmatizer()

print([porter.stem(t) for t in tokens])
print("\n")
print([lancaster.stem(t) for t in tokens])
print("\n")

print([wnl.lemmatize(t) for t in tokens])
```

```python
#Exercise 1a
inputText = input("Input a text :")

tokens = word_tokenize(inputText)

porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
wnl = nltk.WordNetLemmatizer()

print("Porter result: ")
print([porter.stem(t) for t in tokens])
print("\n")
print("Lancaster result: ")
print([lancaster.stem(t) for t in tokens])
print("\n")
print("WordNetLemmatizer result: ")
print([wnl.lemmatize(t) for t in tokens])
print("\n")
```
